```python
#we can convert rgb image to gray scale by converting 3 channels into one channel by taking avergae of 3 channels.
```

```python
gray_image_array = np.mean(img_array, axis=2)
```

```python
gray_image_array.shape
```

```python
#see the third dimention converted into 1
```

```python
plt.imshow(gray_image_array)
```

```python
#lests work on weighted average.
```

```python
img = img_array[:,:,:]
```

```python
img[:,:,0] = 0.2989 * img[:,:,0]
img[:,:,1] = 0.5870 * img[:,:,1]
img[:,:,2] = 0.1140 * img[:,:,2]
```

```python
img = img[:,:,0] + img[:,:,1] + img[:,:,2]
```

```python
img = np.round(img).astype(np.uint8)
```

```python
plt.imshow(img)
```

```python
#lets learn about the binary representation of an image. which is really good in edge detection.
```

```python
#a rgb image can be converted to gray scale and using that gray scale image and a threshold value we can convert into binary representation
```

```python
gray_imag = np.dot(img_array,[0.2989,0.5870,0.1140])
gray_imag
```

```python
#lets consider a threshold value like 128
threshold_value = np.max(gray_imag)*0.5
bin_img = gray_imag>threshold_value
bin_img
```

```python
bin_img = bin_img.astype(np.uint8)*255
```

```python
plt.imshow(bin_img)
```
